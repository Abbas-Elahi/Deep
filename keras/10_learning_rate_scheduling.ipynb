```python
import numpy as np
import plotly.graph_objects as go
```

<h2 dir=rtl align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
زمان‌بندی پله‌ای یا چندضابطه‌ای (Step یا Piecewise)
</font>
</h2>

```python
step = 15
decay = 0.1
epoch = 50
lr = 0.1
x = np.arange(0, epoch, 1)
y = lr * decay ** (x // step)

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines+markers'))
fig.update_layout(
    title="Step Decay Curve",
    xaxis_title="Epoch",
    yaxis_title="Learning Rate",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="#7f7f7f"
    )
)
fig.show()
```

<h2 dir=rtl align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
زمان‌بندی توانی (Power)
</font>
</h2>

```python
lr = 0.1
epoch = 50
power = 1
step = 10

x = np.arange(0, epoch, 1)
# $\alpha_t=\frac{\alpha_0}{(1+{t/s})^c}$
y = lr / (1 + x / step) ** power

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines+markers'))
fig.update_layout(
    title="Power Scheduling Curve",
    xaxis_title="Epoch",
    yaxis_title="Learning Rate",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="#7f7f7f"
    )
)
fig.show()
```

<h2 dir=rtl align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
زمان‌بندی نمایی (Exponential)
</font>
</h2>

```python
lr = 0.1
epoch = 50
step = 15
x = np.arange(0, epoch, 1)
y = lr * 0.1 ** (x / step)

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines+markers'))
fig.update_layout(
    title="Exponential Decay Curve",
    xaxis_title="Epoch",
    yaxis_title="Learning Rate",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="#7f7f7f"
    )
)
fig.show()
```

<h2 dir=rtl align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
زمان‌بندی کسینوسی (Cosine)
</font>
</h2>

```python
lr = 0.1
epoch = 50

x = np.arange(0, epoch, 1)
y = 0.5 * (lr * (1 + np.cos(np.pi * x / epoch)))

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines+markers'))
fig.update_layout(
    title="Cosine Decay Curve",
    xaxis_title="Epoch",
    yaxis_title="Learning Rate",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="#7f7f7f"
    )
)
fig.show()
```

```python

lr = 0.1
epoch = 50
restart = 10
x = np.arange(0, epoch, 1)
y = []
for i in range(epoch):
    if i % restart == 0:
        y.append(lr)
    else:
        y.append(0.5 * (lr * (1 + np.cos(np.pi * (i % restart) / restart))))

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines+markers'))
fig.update_layout(
    title="Cosine Decay Curve with Restart",
    xaxis_title="Epoch",
    yaxis_title="Learning Rate",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="#7f7f7f"
    )
)
fig.show()
```

<h2 dir=rtl align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
زمان‌بندی عملکردی (Performance)
</font>
</h2>

```python
epochs = list(range(1, 30))
validation_loss = [0.5125, 0.6228, 0.4834, 0.5006, 0.6444, 0.5703, 0.6182, 0.6273, 0.3824, 0.3941, 0.4201, 0.4491, 0.4708, 0.456, 0.4131, 0.4473, 0.457, 0.4706, 0.4905, 0.4804, 0.492, 0.5083, 0.5073, 0.5262, 0.5355, 0.5425]

# multiply the learning rate by 0.5 whenever the best validation loss does not improve for five consecutive epochs
learning_rate = [0.02]
best_loss = 1
factor = 0.5
patience = 5
counter = 0
for i in range(len(validation_loss)):
    if validation_loss[i] < best_loss:
        best_loss = validation_loss[i]
        counter = 0
    else:
        counter += 1
    if counter == patience:
        counter = 0
        best_loss = validation_loss[i]
        learning_rate.append(learning_rate[-1] * factor)
    else:
        learning_rate.append(learning_rate[-1])
    

fig = go.Figure()
fig.add_trace(go.Scatter(x=epochs, y=learning_rate[:-1], mode='lines+markers', name='Learning Rate', yaxis='y'))
fig.add_trace(go.Scatter(x=epochs, y=validation_loss, mode='lines+markers', name='Validation Loss', yaxis='y2'))

fig.update_layout(
    title='Learning Rate and Validation Loss',
    xaxis=dict(title='Epoch'),
    yaxis=dict(title='Learning Rate', side='left', rangemode='tozero'),
    yaxis2=dict(title='Validation Loss', side='right', overlaying='y', rangemode='tozero'),
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="#7f7f7f"
    )
)
fig.update_xaxes(dtick=1)
fig.update_layout(legend=dict(
    orientation="h",
    yanchor="bottom",
    y=1.02,
    xanchor="right",
    x=1
))
fig.show()
```
