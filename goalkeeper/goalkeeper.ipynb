<h2 dir=rtl align=right style="line-height:200%;font-family:vazir;color:#0099cc">

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir">
Щ…ШҜЩ„вҖҢЩҮШ§ЫҢ ЫҢШ§ШҜЪҜЫҢШұЫҢ Ш№Щ…ЫҢЩӮ Щ…Ш№Щ…ЩҲЩ„Ш§Щ’ ШҜШ§ШұШ§ЫҢ ШӘШ№ШҜШ§ШҜ ШІЫҢШ§ШҜЫҢ ЩҫШ§ШұШ§Щ…ШӘШұ ЩҮШіШӘЩҶШҜ ЩҲ ШҜШұ ЩҶШӘЫҢШ¬ЩҮ ЩҫШӘШ§ЩҶШіЫҢЩ„ ЩҲ ШӘЩҲШ§ЩҶШ§ЫҢЫҢ ШЁШ§Щ„Ш§ЫҢЫҢ Ш¬ЩҮШӘ ЫҢШ§ШҜЪҜЫҢШұЫҢ ШҜШ§ШұЩҶШҜ. ЪҶЩҶЫҢЩҶ ЩҫШӘШ§ЩҶШіЫҢЩ„ЫҢ Ш¬ЩҮШӘ ЩҲЩӮЩҲШ№ ЩҶЫҢШ§ШІ ШЁЩҮ ШӘШ№ШҜШ§ШҜ ШІЫҢШ§ШҜЫҢ ШҜШ§ШҜЩҮвҖҢЫҢ ШўЩ…ЩҲШІШҙЫҢ ШҜШ§ШұШҜ ЩҲ Ш§ЪҜШұ ШӯШ¬Щ… ШҜШ§ШҜЩҮвҖҢЩҮШ§ЫҢ ШҜШұ ШҜШіШӘШұШі Щ…ШҜЩ„ Ъ©Щ… ШЁШ§ШҙШҜ ШЁЩҮвҖҢШ§ШӯШӘЩ…Ш§Щ„ ШІЫҢШ§ШҜ Щ…ШҜЩ„ ШЁШұ ШұЩҲЫҢ ЩҮЩ…Ш§ЩҶвҖҢЩҮШ§ ШЁЫҢШҙвҖҢШЁШұШ§ШІШҙ Щ…ЫҢвҖҢШҙЩҲШҜ. ЫҢШ№ЩҶЫҢ Щ…ШҜЩ„ ШЁШұ ШұЩҲЫҢ ШҜШ§ШҜЩҮвҖҢЩҮШ§ЫҢ ШўЩ…ЩҲШІШҙЫҢ ШЁШіЫҢШ§Шұ Ш®ЩҲШЁ Ш№Щ…Щ„ Щ…ЫҢвҖҢЪ©ЩҶШҜШҢ Ш§Щ…Ш§ ЩӮШ§ШҜШұ ЩҶЫҢШіШӘ ЪҶЫҢШІЩҮШ§ЫҢЫҢ Ъ©ЩҮ ЫҢШ§ШҜ ЪҜШұЩҒШӘЩҮ Ш§ШіШӘ ШұШ§ ШЁЩҮ ЩҶЩ…ЩҲЩҶЩҮвҖҢЩҮШ§ЫҢ Ш¬ШҜЫҢШҜ ЩҲ ШӘШ§Ъ©ЩҶЩҲЩҶ ШҜЫҢШҜЩҮвҖҢЩҶШҙШҜЩҮ ШӘШ№Щ…ЫҢЩ… ШҜЩҮШҜ. ШҜШұ Ш§ЫҢЩҶ ШӘЩ…ШұЫҢЩҶ ШЁЩҮвҖҢШҙЪ©Щ„ Ш№Щ…Щ„ЫҢ ЪҶЩҶЫҢЩҶ Щ…ЩҲШ¶ЩҲШ№ЫҢ ШұШ§ Щ…ШҙШ§ЩҮШҜЩҮ Ш®ЩҲШ§ЩҮЫҢЩ… Ъ©ШұШҜ ЩҲ ЪҶЩҶШҜ ШӘЪ©ЩҶЫҢЪ© ШұЪҜЩҲЩ„Ш§ШұЫҢШІЫҢШҙЩҶ ШұШ§ ШЁШұШ§ЫҢ Ш¬Щ„ЩҲЪҜЫҢШұЫҢ Ш§ШІ ШЁЫҢШҙвҖҢШЁШұШ§ШІШҙ Щ…ШҜЩ„ ШЁЩҮ Ъ©Ш§Шұ Ш®ЩҲШ§ЩҮЫҢЩ… ШЁШұШҜ.
<br>
Щ…ШіШҰЩ„ЩҮвҖҢЫҢ Щ…Ш§ Ш§ЫҢЩҶ Ш§ШіШӘ Ъ©ЩҮ ШҜШұЩҲШ§ШІЩҮвҖҢШЁШ§ЩҶ ШӘЫҢЩ… Щ…Щ„ЫҢ ЩҒШұШ§ЩҶШіЩҮ Щ…ЫҢвҖҢШ®ЩҲШ§ЩҮШҜ ШӘЩҲЩҫ ШұШ§ ШЁЩҮ ЩҲШіШ· ШІЩ…ЫҢЩҶ ЩҫШұШӘШ§ШЁ Ъ©ЩҶШҜ ШӘШ§ ЫҢЪ©ЫҢ Ш§ШІ ШЁШ§ШІЫҢЪ©ЩҶШ§ЩҶ Ш®ЩҲШҜЫҢ ШЁШӘЩҲШ§ЩҶШҜ ШЁШ§ ШіШұ ШЁЩҮ ШўЩҶ Ш¶ШұШЁЩҮ ШЁШІЩҶШҜ. ШЁЩҮвҖҢШ№ЩҶЩҲШ§ЩҶ Щ…Ш«Ш§Щ„ ШҜШұ ШҙЪ©Щ„ ШІЫҢШұ ЫҢЪ© ШІЩ…ЫҢЩҶ ЩҒЩҲШӘШЁШ§Щ„ ШұШ§ Щ…ШҙШ§ЩҮШҜЩҮ Щ…ЫҢвҖҢЪ©ЩҶЫҢШҜ Ъ©ЩҮ ЩҶШӯЩҲЩҮвҖҢЫҢ ЩҫШұШӘШ§ШЁ ШӘЩҲЩҫ ШӘЩҲШіШ· ШҜШұЩҲШ§ШІЩҮвҖҢШЁШ§ЩҶ ЩҲ Щ…ЩҲЩӮШ№ЫҢШӘ ШЁШ§ШІЫҢЪ©ЩҶШ§ЩҶ ШұШ§ ЩҶШҙШ§ЩҶ Щ…ЫҢвҖҢШҜЩҮШҜ. Ш§Ш·Щ„Ш§Ш№Ш§ШӘ ШӘЩ…Ш§Щ…ЫҢ ЩҫШұШӘШ§ШЁвҖҢЩҮШ§ ШҜШұ ЫұЫ° ШЁШ§ШІЫҢ ЪҜШ°ШҙШӘЩҮвҖҢЫҢ ШӘЫҢЩ… Щ…Щ„ЫҢ ЩҒШұШ§ЩҶШіЩҮ ШҜШұ Ш§Ш®ШӘЫҢШ§Шұ ШҙЩ…Ш§ ЩӮШұШ§Шұ ЪҜШұЩҒШӘЩҮ ЩҲ ЩӮШөШҜ ШҜШ§ШұЫҢЩ… Щ…ШҜЩ„ЫҢ ШўЩ…ЩҲШІШҙ ШҜЩҮЫҢЩ… Ъ©ЩҮ ШЁШӘЩҲШ§ЩҶШҜ ЩҫЫҢШҙвҖҢШЁЫҢЩҶЫҢ Ъ©ЩҶШҜ ШўЫҢШ§ ШӘЩҲЩҫ ШЁЩҮ ШЁШ§ШІЫҢЪ©ЩҶ Ш®ЩҲШҜЫҢ Щ…ЫҢвҖҢШұШіШҜ ЫҢШ§ Ш®ЫҢШұ. ШҜШұ ЩҶЩҮШ§ЫҢШӘ ШЁШ§ Ш§ШіШӘЩҒШ§ШҜЩҮ Ш§ШІ Щ…ШҜЩ„ Щ…ЫҢвҖҢШӘЩҲШ§ЩҶЫҢЩ… ШҜШұ ШӘЩ…ШұЫҢЩҶвҖҢЩҮШ§ЫҢ ШўЫҢЩҶШҜЩҮ ЩҫШұШӘШ§ШЁвҖҢЩҮШ§ЫҢ ШҜШұЩҲШ§ШІЩҮвҖҢШЁШ§ЩҶ ШұШ§ ШЁЩҮШЁЩҲШҜ ШЁШ®ШҙЫҢЩ….

```python
def plot_loss_acc(history):
    train_loss = history.history['loss']
    val_loss = history.history['val_loss']
    train_acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    
    epochs = range(1, len(train_loss) + 1)
    
    fig, axs = plt.subplots(2, figsize=(10, 7))
    fig.suptitle('Training and Validation Metrics')
    
    # Loss plot
    axs[0].plot(epochs, train_loss, label='Training loss', c='red')
    axs[0].plot(epochs, val_loss, label='Validation loss', c='blue')
    axs[0].set_title('Loss')
    axs[0].set_xlabel('Epochs')
    axs[0].set_ylabel('Loss')
    axs[0].legend()
    axs[0].set_ylim([0, 1.5])
    
    # Accuracy plot
    axs[1].plot(epochs, train_acc, label='Training accuracy', c='red')
    axs[1].plot(epochs, val_acc, label='Validation accuracy', c='blue')
    axs[1].set_title('Accuracy')
    axs[1].set_xlabel('Epochs')
    axs[1].set_ylabel('Accuracy')
    axs[1].legend()
    
    plt.subplots_adjust(hspace=0.5)
    plt.show()
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay

def get_decision_boundaries(model, xmin, xmax, ymin, ymax, steps):
  x_span = np.linspace(xmin, xmax, steps)
  y_span = np.linspace(ymin, ymax, steps)
  xx, yy = np.meshgrid(x_span, y_span)
  points = (np.stack([xx.ravel(), yy.ravel()], axis=1).astype(np.float32))

  z = (model.predict(np.c_[xx.ravel(), yy.ravel()])>0.5).reshape(xx.shape)

  return xx, yy, z

def plt_decision_boundaries(model, x_min, x_max, y_min, y_max, steps):
  plt.figure(figsize=(6, 4))
  xx, yy, z = get_decision_boundaries(model, x_min, x_max, y_min, y_max, steps)
  plt.contourf(xx, yy, z, alpha=0.2, cmap=matplotlib.colors.ListedColormap(["C1", "C0"]));
```

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir">
Ш§ЫҢЩҶвҖҢШЁШ§Шұ Щ…ЫҢвҖҢШ®ЩҲШ§ЩҮЫҢЩ… ШЁЩҮвҖҢШ¬Ш§ЫҢ Ш§ШіШӘЩҒШ§ШҜЩҮ Ш§ШІ ШӯШ°ЩҒ ШӘШөШ§ШҜЩҒЫҢШҢ Ш§ШІ ШұЪҜЩҲЩ„Ш§ЫҢШІШұ <code>L1L2</code> ШЁШұШ§ЫҢ ЩҲШІЩҶвҖҢЩҮШ§ЫҢ Ъ©ШұЩҶЩ„ ЩҮШұ Щ„Ш§ЫҢЩҮ Ш§ШіШӘЩҒШ§ШҜЩҮ Ъ©ЩҶЫҢЩ…. ШЁШұШ§ЫҢ Щ…Ш·Ш§Щ„Ш№ЩҮвҖҢЫҢ Щ…ШіШӘЩҶШҜШ§ШӘ Ш§ЫҢЩҶ ЩҶЩҲШ№ ШұЪҜЩҲЩ„Ш§ШұЫҢШІЫҢШҙЩҶ Щ…ЫҢвҖҢШӘЩҲШ§ЩҶЫҢШҜ ШЁЩҮ <a href="https://keras.io/api/layers/regularizers/#l1l2-class" target="_blank">Ш§ЫҢЩҶ Щ„ЫҢЩҶЪ©</a> Щ…ШұШ§Ш¬Ш№ЩҮ Ъ©ЩҶЫҢШҜ. Ш¶ШұЫҢШЁ <code>l1</code> ШұШ§ Щ…Ш№Ш§ШҜЩ„ ШЁШ§ <code>3e-5</code> ЩҲ Ш¶ШұЫҢШЁ <code>l2</code> ШұШ§ Щ…Ш№Ш§ШҜЩ„ ШЁШ§ <code>3e-4</code> ЩӮШұШ§Шұ ШҜЩҮЫҢШҜ.
</font>
</p>

```python
model_l1l2 = keras.Sequential([
    keras.layers.Input(shape=(2,)),
    keras.layers.Dense(5000, activation='relu',
                       kernel_regularizer=keras.regularizers.L1L2(l1=3e-5, l2=3e-4)),
    keras.layers.Dense(5000, activation='relu',
                       kernel_regularizer=keras.regularizers.L1L2(l1=3e-5, l2=3e-4)),
    keras.layers.Dense(5000, activation='relu',
                       kernel_regularizer=keras.regularizers.L1L2(l1=3e-5, l2=3e-4)),
    keras.layers.Dense(1, activation='sigmoid',
                       kernel_regularizer=keras.regularizers.L1L2(l1=3e-5, l2=3e-4))
])
```

    ШЁШұШ§ЫҢ ШіШ§Ш®ШӘЩҮвҖҢШҙШҜЩҶ ЩҒШ§ЫҢЩ„ <code>result.zip</code> ШіЩ„ЩҲЩ„вҖҢЩҮШ§ЫҢ ШІЫҢШұ ШұШ§ Ш§Ш¬ШұШ§ Ъ©ЩҶЫҢШҜ. ШӘЩҲШ¬ЩҮ ШҜШ§ШҙШӘЩҮ ШЁШ§ШҙЫҢШҜ Ъ©ЩҮ ЩҫЫҢШҙ Ш§ШІ Ш§Ш¬ШұШ§ЫҢ ШіЩ„ЩҲЩ„ ШІЫҢШұ ШӘШәЫҢЫҢШұШ§ШӘ Ш§Ш№Щ…Ш§Щ„ ШҙШҜЩҮ ШҜШұ ЩҶШӘвҖҢШЁЩҲЪ© ШұШ§ Ш°Ш®ЫҢШұЩҮ Ъ©ШұШҜЩҮ ШЁШ§ШҙЫҢШҜ (<code>ctrl+s</code>) ШӘШ§ ШҜШұ ШөЩҲШұШӘ ЩҶЫҢШ§ШІ ШЁЩҮ ЩҫШҙШӘЫҢШЁШ§ЩҶЫҢ Ш§Щ…Ъ©Ш§ЩҶ ШЁШұШұШіЫҢ Ъ©ШҜ ШҙЩ…Ш§ ЩҲШ¬ЩҲШҜ ШҜШ§ШҙШӘЩҮ ШЁШ§ШҙШҜ. ЩҮЩ…ЪҶЩҶЫҢЩҶ Ш§ЪҜШұ Ш§ШІ ЪҜЩҲЪҜЩ„ Ъ©ЩҲЩ„ШЁ Ш§ШіШӘЩҒШ§ШҜЩҮ Щ…ЫҢвҖҢЪ©ЩҶЫҢШҜШҢ ШҜШұ ШөЩҲШұШӘ ЩҶЫҢШ§ШІ ШЁЩҮ ЩҫШҙШӘЫҢШЁШ§ЩҶЫҢ ШӯШӘЩ…Ш§ЩӢ ШўШ®ШұЫҢЩҶ ЩҶШіШ®ЩҮ Ш§ШІ ЩҶШӘвҖҢШЁЩҲЪ© ШұШ§ ШЁЩҮвҖҢШөЩҲШұШӘ ШҜШіШӘЫҢ ШҜШ§ЩҶЩ„ЩҲШҜ Ъ©ШұШҜЩҮ ЩҲ ШҜШ§Ш®Щ„ ЩҒШ§ЫҢЩ„ Ш§ШұШіШ§Щ„ЫҢ ЩӮШұШ§Шұ ШҜЩҮЫҢШҜ ЫҢШ§ Щ„ЫҢЩҶЪ© Ъ©ЩҲЩ„ШЁ ШұШ§ ШЁШ§ Щ…Ш§ ШЁЩҮвҖҢШ§ШҙШӘШұШ§Ъ© ШЁЪҜШ°Ш§ШұЫҢШҜ.
</font>

```python
import numpy as np
import json

def get_kernel_regularizer_info(layer):
  try:
      con = layer.get_config()['kernel_regularizer']
      con['config']['l1']=round(con['config']['l1'], 6)
      con['config']['l2']=round(con['config']['l2'], 6)
      return str(con)
  except:
      return ""

model_l1l2_info=[]
for layer in model_l1l2.layers:
  if layer.__class__.__name__=="Dense":
    model_l1l2_info.append({
        "name": layer.__class__.__name__,
        "units":layer.units,
        "activation":layer.get_config()["activation"],
        "kernel_regularizer": get_kernel_regularizer_info(layer)
    })
  elif layer.__class__.__name__=="Dropout":
    model_l1l2_info.append({
        "name": layer.__class__.__name__,
        "trainable":layer.get_config()["trainable"],
        "rate": layer.get_config()["rate"],
    })
  else:
    model_l1l2_info.append({"name": layer.__class__.__name__})

with open("model_l1l2_info.json", "w") as f:
    json.dump(model_l1l2_info, f)

model_dropout_info=[]
for layer in model_dropout.layers:
  if layer.__class__.__name__=="Dense":
    model_dropout_info.append({
        "name": layer.__class__.__name__,
        "units":layer.units,
        "activation":layer.get_config()["activation"],
        "kernel_regularizer": get_kernel_regularizer_info(layer)
    })
  elif layer.__class__.__name__=="Dropout":
    model_dropout_info.append({
        "name": layer.__class__.__name__,
        "trainable":layer.get_config()["trainable"],
        "rate": layer.get_config()["rate"],
    })
  else:
    model_dropout_info.append({"name": layer.__class__.__name__})

with open("model_dropout_info.json", "w") as f:
    json.dump(model_dropout_info, f)
```
